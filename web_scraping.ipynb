````python
def process_people(person):
    name = person.find('h2').text
    quote = person.find_all('p')[0].text.strip()
    email = person.find_all('p')[1].text
    phone_and_ext = person.find_all('p')[2].text
    address = person.find_all('p')[3].text.strip()
    
    return {
        "name": name,
        "quote": quote,
        "email": email,
        "phone_and_ext": phone,
        "address": address
    }
````

````python
def codeup_blogs(urls):
    blog_df = pd.DataFrame(columns=['title', 'body'])
    for url in urls:
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
        response = requests.get(url, headers=headers)
        html = response.text
        soup = bs4.BeautifulSoup(html)
        container = soup.select('.container')[0]
        title = container.find('h1').text
        body = container.find(itemprop='articleBody').text
        container_dict = {
                    'title': title,
                    'body': body
                        }
        container_df = pd.DataFrame(container_dict,index=[0])
        blog_df = pd.concat([blog_df, container_df], axis=0)
        
    return blog_df.drop_duplicates().reset_index(drop=True)
````

````python
def codeup_blogs(urls):
    blog_df = pd.DataFrame(columns=['title', 'body'])
    for url in urls:
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
        response = requests.get(url, headers=headers)
        html = response.text
        soup = bs4.BeautifulSoup(html)
        container = soup.select('.container')[0]
        title = container.find('h1').text
        body = container.find(itemprop='articleBody').text
        container_dict = {
                    'title': title,
                    'body': body
                        }
        container_df = pd.DataFrame(container_dict,index=[0])
        blog_df = pd.concat([blog_df, container_df], axis=0)
        
    return blog_df.drop_duplicates().reset_index(drop=True)

def all_codeup_blogs():
    url = 'https://codeup.com/blog/'
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    response = requests.get(url, headers=headers)
    html = response.text
    soup = bs4.BeautifulSoup(html)
    
    urls = []
    for a in soup.select('a.jet-listing-dynamic-link__link', href=True):
        urls.append(a['href'])
        urls = list(set(urls))
    
    blogs_df = codeup_blogs(urls)
    
    return blogs_df
````

````python
url = ('https://inshorts.com/en/read/business')
response = requests.get(url)
html = response.text
soup = bs4.BeautifulSoup(html)
````

````python
business_container = soup.select('.card-stack')[0]
business_container
````

````python
business_articles = business_container.select(".news-card.z-depth-1")
````

````python
business_article = business_articles[0]
````

````python
def inshorts_articles(business_article):
    return {
        'title': business_article.find(itemprop="headline").text,
        'body': business_article.find(itemprop="articleBody").text,
        'author': business_article.find(class_="author").text,
        'date_modified': business_article.find(clas="date").text,
        'time_modified': business_article.find(class_="time").text,
    }
````

````python
business_df = pd.DataFrame([inshorts_articles(business_article) for business_article in business_articles])
````
